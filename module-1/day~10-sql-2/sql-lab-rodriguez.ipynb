1. List all customer names and their addresses who don't have a specified sales rep. Order by customer name.

2. List the names of all customers that have a creditLimit between 100,000 and 200,000 inclusively. Order by customer name.

```python
pd.read_sql_query(""" SELECT customerName, creditLimit
                      FROM customers
                      WHERE creditLimit BETWEEN 100000 AND 200000
                      ORDER BY 1 ASC;""", con)
```


3. List all employees who have the initials of M.P. or P.M., by name. Order by last name.

```python
pd.read_sql_query(""" SELECT firstName, lastName
                      FROM employees
                      WHERE firstName LIKE 'm%' AND lastName LIKE 'p%' or firstName LIKE 'p%' AND lastName LIKE 'm%'
                      ORDER BY 2;""", con)
```

4. List all the products' productNames purchased by Mini Wheels Co. Order the list by product line.

```python
pd.read_sql_query(""" SELECT productName, productLine, customerName
                      FROM products
                      JOIN orderdetails USING(productCode)
                      JOIN orders USING(orderNumber)
                      JOIN customers USING(customerNumber)
                      WHERE customerName ='Mini Wheels Co.'
                      GROUP BY productLine, productName;""", con)
```

5. List the customers by customerName whose contact person shares either a first name or last name with any of the company's employees Order by customerName

```python
pd.read_sql_query(""" SELECT contactFirstName , contactLastName, firstName, lastName
                      FROM customers c
                      JOIN employees e
                      ON c.salesRepEmployeeNumber = e.employeeNumber
                      WHERE c.contactFirstName = e.firstName OR c.contactLastName = e.lastName """, con)
```

9. Identify the offices (by city) that have the least employees and the most employees (answer should include city, numEmps)

```python
pd.read_sql_query(""" SELECT city, count(employeeNumber)
                      FROM offices
                      JOIN employees
                      USING(officeCode)
                      GROUP BY city
                      ORDER BY count(employeeNumber) """, con)
```

```python
pd.read_sql_query(""" SELECT city, count(*) as employeeCount
                      FROM employees
                      JOIN offices USING (officeCode)
                      GROUP BY officeCode, city
                      
                      HAVING employeeCount in (SELECT max(employeeCount)
                                               FROM (SELECT officeCode, count(*) as employeeCount FROM employees GROUP BY officeCode) AS max_count
                                               UNION
                                               SELECT min(employeeCount)
                                               FROM (SELECT officeCode, count(*) as employeeCount FROM employees GROUP BY officeCode) AS min_count
                                               ) """, con)
```

10. For all Sales Reps list their names (first and last) and the name of the office they work out of and the number of clients they represent. Order by reverse order of numClients

```python
pd.read_sql_query(""" SELECT firstName, lastName, o.city, COUNT(c.salesRepEmployeeNumber) as num_clients
                      FROM employees e
                      JOIN offices o
                      USING(officeCode)
                      JOIN customers c
                      ON e.employeeNumber = c.salesRepEmployeeNumber
                      GROUP BY salesRepEmployeeNumber
                      ORDER BY COUNT(c.salesRepEmployeeNumber) DESC """, con)
```

11. List the customer(s) and their total payments (answer should include customerNumber, customerName, totalPayment where totalPayment is the total amount of dollars the customer has paid). Order by totalPayment.

```python
pd.read_sql_query(""" SELECT customerNumber, customerName, SUM(amount) as totalPayment
                      FROM customers
                      JOIN payments
                      USING(customerNumber)
                      GROUP BY customerNumber
                      ORDER BY totalPayment DESC """, con)
```

14. List the average order size for each customer (answer should include customerName, avgQuantity) Order by customer name.

```python
pd.read_sql_query(""" SELECT customerName , AVG(quantityOrdered)
                      FROM customers c
                      JOIN orders
                      USING(customerNumber)
                      JOIN orderdetails
                      USING(orderNumber)
                      GROUP BY customerName
                      ORDER BY AVG(quantityOrdered) DESC """, con)
```
